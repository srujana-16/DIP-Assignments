## Part 1

```python
# Takes in im - an 8-bit image and k - the bit-plane number where 1 ≤ k ≤ 8. The function then returns the kth bit-plane of the image.
def bitSlice(im,k):

  # Extract the heights and width of the image
  im = im.copy()
  h,w = im.shape
  for i in range(h):
    for j in range(w):
        # temp = ((im[i][j])*(2**(8-k)))%2
        im[i][j] = ((im[i][j]) >> (8-k))%2
  return im 
```

# Part 2

```python
# All the bit planes for rupee.jpg, securityThreat.jpg and pepper.jpg.
rupee = cv2.imread('rupee.jpg')
security = cv2.imread('securityThreat.jpg')
pepper = cv2.imread('pepper.jpg')

# Conveting to grayscale
rupee = cv2.cvtColor(rupee, cv2.COLOR_BGR2GRAY)
security = cv2.cvtColor(security, cv2.COLOR_BGR2GRAY)
pepper = cv2.cvtColor(pepper, cv2.COLOR_BGR2GRAY)
imgs = [rupee, security, pepper]

# PLotting 
for im in imgs:
    plt.figure(figsize = (20,10))
    for j in range(1,9):
        plt.subplot(2,4,j)
        output = bitSlice(im, j)
        plt.imshow(output, cmap='gray')
        plt.title('Bit Plane {}'.format(j))
        plt.axis('off')
    plt.show()  
```

# Part 3

```python
def bitQuantizer(img,k):
    return np.around(img*((2**k-1)/255)).astype(int)
```

# Part 4

```python
# All the bit planes for rupee.jpg, securityThreat.jpg and pepper.jpg.
rupee = cv2.imread('rupee.jpg')
security = cv2.imread('securityThreat.jpg')
pepper = cv2.imread('pepper.jpg')

# Conveting to grayscale
rupee = cv2.cvtColor(rupee, cv2.COLOR_BGR2GRAY)
security = cv2.cvtColor(security, cv2.COLOR_BGR2GRAY)
pepper = cv2.cvtColor(pepper, cv2.COLOR_BGR2GRAY)
imgs = [rupee, security, pepper]

# PLotting 
for im in imgs:
    plt.figure(figsize = (20,10))
    for j in range(1,9):
        plt.subplot(2,4,j)
        output = bitQuantizer(im, j)
        plt.imshow(output, cmap='gray')
        plt.title('{} Bit Quantisation'.format(j))
        plt.axis('off')
    plt.show() 
```

### - Bit plane slicing is a method of representing an image with one or more bits of the byte used for each pixel. One can use only MSB to represent the pixel, which reduces the original gray level to a binary image.
### - In the quantization results we see that a k-bit quantized image contains the information from the  1st  to  kth  bit planes, hence, the 8-bit quantized image is the combination of all the bits.
